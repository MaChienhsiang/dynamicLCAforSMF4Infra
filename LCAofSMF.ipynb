```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
EI_bld = 188.81 # impact of the factory building in ton CO2-eq
EI_mac = 174.39 # impact of the machinery in ton CO2-eq
EI_sys = 20.97 # impact of the building service system of the SMF, such as the lighting, air conditioner, etc.in ton CO2-eq
EI_p = (EI_bld+EI_mac+EI_sys) # the total impact in the production stage in ton CO2-eq
```

```python
cate = ['Production Stage', 'Construction Stage','Usage Stage']
EI_bld = np.array([188.81, 0,0])
EI_mac = np.array([174.39, 0,0])
EI_sys = np.array([20.97, 0,0])
cons = np.array([0, 39.25,0])
trans = np.array([0, 1.4,0])
EI_u = np.array([0, 0,9067])
# plot bars in stack manner
#plt.figure(dpi=300)
plt.bar(cate, EI_bld, label='building')
plt.bar(cate, EI_mac, bottom=EI_bld, label='machinery')
plt.bar(cate, EI_sys, bottom= EI_mac+EI_bld, label='service')
plt.bar(cate, cons, label='construction')
plt.bar(cate, trans, bottom=cons, label='transport')
plt.bar(cate, EI_u, label='tube production')
plt.ylim(0,9200)
plt.ylabel('GHG emission [ton CO$_2$-eq]')
plt.legend()
plt.show()
```
